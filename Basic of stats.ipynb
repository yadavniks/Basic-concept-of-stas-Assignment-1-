```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Skewness:The distribution appears to be right-skewed (positively skewed).
#This is evident from the long tail extending towards higher volume values (15–30), while the majority of the data is concentrated between 0 and 10.
#In a right-skewed distribution, the mean is typically greater than the median, and the bulk of the data lies on the left side of the histogram.
```

```python
from sklearn.preprocessing import StandardScaler
# Selecting numerical columns for standardization
numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
numerical_columns
```

```python
scaler = StandardScaler()
scaler
```

```python
df.standardized =df. copy()  # Creating a copy of the original dataset
df.standardized[numerical_columns] = scaler.fit_transform (df[numerical_columns])
df.standardized[numerical_columns] 
```

```python
df.standardized.describe()
```

```python
for column in numerical_columns:
    plt.figure(figsize=(10, 4))

    # Original Data Distribution
    plt.subplot(1, 2, 1)
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {column} (Original)')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    # Standardized Data Distribution
    plt.subplot(1, 2, 2)
    plt.hist(df.standardized[column], bins=20, color='salmon', edgecolor='black')
    plt.title(f'Distribution of {column} (After Standardization)')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()
```
